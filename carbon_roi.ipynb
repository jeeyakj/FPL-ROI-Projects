```python
def calculate_roi(investment, return_amount):
    return ((return_amount - investment) / investment) * 100

def carbon_adjusted_roi(investment, return_amount, carbon_emissions, offset_cost_per_ton):
    roi = calculate_roi(investment, return_amount)
    carbon_penalty = carbon_emissions * offset_cost_per_ton
    adjusted_roi = roi - carbon_penalty
    return adjusted_roi
```

```python
companies = [
    {"name": "GreenTech Ltd", "investment": 1000, "return": 1300, "carbon_emissions": 1.2},
    {"name": "OilCorp", "investment": 1000, "return": 1500, "carbon_emissions": 5.0}
]
```

```python
import matplotlib.pyplot as plt

names = [c["name"] for c in companies]
rois = [calculate_roi(c["investment"], c["return"]) for c in companies]
carbon_rois = [carbon_adjusted_roi(c["investment"], c["return"], c["carbon_emissions"], 20) for c in companies]

plt.bar(names, rois, label="ROI")
plt.bar(names, carbon_rois, label="Carbon ROI", alpha=0.7)
plt.legend()
plt.title("ROI vs Carbon-Adjusted ROI")
plt.ylabel("ROI (%)")
plt.show()
```
